```python
import pandas as pd
```

```python
for country in ['Egypt', 'Hungary', 'Nigeria', 'Poland', 'Romania']:
    
    gdp_path = 'FER_GDP_ratio_' + country + '.csv'
    gdp = pd.read_csv(gdp_path)
    gdp['FER/GDP MA3'] = gdp['FER/GDP'].rolling(3).mean()
    csv_path = 'FER_GDP_ratio_' + country + '.csv'
    gdp.to_csv(csv_path, index = False)
```
