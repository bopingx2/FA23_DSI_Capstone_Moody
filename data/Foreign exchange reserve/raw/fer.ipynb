```python
for country in ['Egypt', 'Hungary', 'Nigeria', 'Poland', 'Romania']:
    fer_path = 'Foreign Exchange Reserves USD mn Monthly ' + country + '.csv'
    usd = pd.read_csv(fer_path)
    fer = usd.iloc[26:]
    usd_name = 'Foreign Exchange Reserves: USD mn: Monthly: ' + country
    fer.rename(columns={'Unnamed: 0': 'Date', usd_name: 'Foreign Exchange Reserves USD mn'}, inplace = True)
    fer['Date'] = pd.to_datetime(fer['Date']) + pd.tseries.offsets.MonthEnd(0)
    fer['Foreign Exchange Reserves USD mn'] = fer['Foreign Exchange Reserves USD mn'].astype(float)

    csv_path = 'FER_in_mil_USD_' + country + '.csv'
    fer.to_csv(csv_path, index = False)
```
