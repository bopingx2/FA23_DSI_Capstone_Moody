```python
import pandas as pd
```

```python
for country in ['Egypt', 'Hungary', 'Nigeria', 'Poland', 'Romania']:
    lend_path = 'Foreign Exchange Reserves USD mn Monthly ' + country + '.csv'
    lend = pd.read_csv(lend_path)
    lend = lend.iloc[26:]
    lend_name = 'Total Loans: YoY: Monthly: ' + country
    lend.rename(columns={'Unnamed: 0': 'Date', lend_name: 'Bank Lending yryr%'}, inplace = True)
    lend['Date'] = pd.to_datetime(lend['Date']) + pd.tseries.offsets.MonthEnd(0)
    lend['Foreign Exchange Reserves USD mn'] = lend['Foreign Exchange Reserves USD mn'].astype(float)

    csv_path = 'FER_in_mil_USD_' + country + '.csv'
    lend.to_csv(csv_path, index = False)
```
